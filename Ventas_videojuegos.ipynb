```python
console = Videogames['Platform'].value_counts()
```

```python
plt.figure(figsize=(18, 6))
sns.barplot(x=console.index, y=console.values, palette='rocket')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='Platform', y='Global_Sales', data=Videogames)
plt.title('distribucion')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='Genre', y='Global_Sales', data=Videogames)
plt.title('distribucion')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='Platform', y='Genre', data=Videogames)
plt.title('distribucion')
plt.show()
```

```python
plt.figure(figsize=(15, 105))
sns.boxplot(x='Genre', y='Publisher', data=Videogames)
plt.title('distribucion')
plt.show()
```

```python
Genre_counts = Videogames['Genre'].value_counts()

plt.figure(figsize=(12,6))
ax = sns.barplot(x=Genre_counts.index, y=Genre_counts.values, palette = 'viridis')
plt.title('cantidad de juegos por genero')
plt.xlabel('generos')
plt.ylabel('cantidad')
plt.xticks(rotation=45)
plt.show()
```

```python
Top_Publishers = Videogames[['Publisher', 'Global_Sales']].sort_values(by='Global_Sales', ascending=False).head(70) #definiendo el grupo
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x='Publisher', y='Global_Sales', data=Top_Publishers, palette='inferno')
plt.title('los 10 con mas ventas')
plt.xlabel('editor')
plt.ylabel('ventas')
plt.xticks(rotation=90)
plt.show()
```

```python
Top_Games = Videogames[['Name', 'Global_Sales']].sort_values(by='Global_Sales', ascending=False).head(20) #definiendo el grupo
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x='Name', y='Global_Sales', data=Top_Games, palette='inferno')
plt.title('los 10 con mas ventas')
plt.xlabel('juego')
plt.ylabel('ventas')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(x='Year', y='Global_Sales', data=Videogames, palette='inferno')
plt.title('ventas por ao')
plt.show()
```

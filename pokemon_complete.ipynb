```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi



# Elegir el Pokémon específico (ejemplo: Pikachu)
pokemon_name = 'Zapdos'  # Cambia esto por el nombre del Pokémon que quieras analizar
pokemon_data = pokemon[pokemon['Name'] == pokemon_name]

# Verificar si el Pokémon existe en el dataset
if not pokemon_data.empty:
    # Extraer las estadísticas del Pokémon
    pokemon_stats = pokemon_data[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].values.flatten()

    # Definir las categorías (estadísticas)
    categories = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
    N = len(categories)

    # Crear el ángulo para cada categoría en el gráfico de radar
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]  # Cerrar el gráfico (mismo ángulo inicial y final)

    # Añadir el primer valor de stats al final para cerrar el gráfico
    stats = list(pokemon_stats)
    stats += stats[:1]  # Repetimos el primer valor para cerrar el gráfico

    # Iniciar el gráfico en formato polar
    plt.figure(figsize=(6, 6))
    ax = plt.subplot(111, polar=True)

    # Primer punto en el gráfico de radar
    plt.xticks(angles[:-1], categories)

    # Añadir las estadísticas del Pokémon al gráfico
    ax.plot(angles, stats, linewidth=2, linestyle='solid')

    # Rellenar el área debajo de la curva
    ax.fill(angles, stats, 'b', alpha=0.3)

    # Título del gráfico
    plt.title(f'Estadísticas de {pokemon_name}', size=15, color='blue', y=1.1)

    # Mostrar el gráfico
    plt.show()
else:
    print(f"El Pokémon {pokemon_name} no existe en el dataset.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python

# Variables predictoras (estadísticas)
X = pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Variable objetivo (Tipo de Pokémon)
y = pokemon['Type 1']  # Si 'Type 1' es la columna que contiene el tipo principal de Pokémon

# Dividir en datos de entrenamiento y prueba (70% para entrenamiento y 30% para prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo de Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Precisión del modelo: {accuracy:.2f}')

# Mostrar un reporte de clasificación
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Definir el espacio de búsqueda de hiperparámetros
param_grid = {
    'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10]
            }

# Realizar la búsqueda de hiperparámetros
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3)
grid_search.fit(X_train, y_train)

# Mejor combinación de hiperparámetros
print(f"Mejores hiperparámetros: {grid_search.best_params_}")

# Evaluar el modelo con los mejores hiperparámetros
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
print(f'Precisión del modelo ajustado: {accuracy_score(y_test, y_pred):.2f}')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python

# Variables predictoras (estadísticas)
X = pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Variable objetivo (Tipo de Pokémon)
y = pokemon['Type 1']  # La columna que contiene el tipo de Pokémon

# Estandarizar las variables predictoras
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir en datos de entrenamiento y prueba (70% para entrenamiento y 30% para prueba)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Crear el modelo de SVM con kernel lineal
model = SVC(kernel='linear', random_state=42)

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Precisión del modelo: {accuracy:.2f}')

# Mostrar un reporte de clasificación
print(classification_report(y_test, y_pred))
```

```python
# Crear el modelo de SVM con kernel RBF
model_rbf = SVC(kernel='rbf', random_state=42)

# Entrenar el modelo
model_rbf.fit(X_train, y_train)

# Predecir con el modelo de kernel RBF
y_pred_rbf = model_rbf.predict(X_test)

# Evaluar el modelo con kernel RBF
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
print(f'Precisión del modelo con kernel RBF: {accuracy_rbf:.2f}')

# Mostrar reporte de clasificación
print(classification_report(y_test, y_pred_rbf))
```
